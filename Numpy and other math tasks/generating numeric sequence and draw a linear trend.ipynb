(EN)  
Write a script in which:
- generate a random numerical sequence of 10 elements (from 1 to 20)
- plot a graph of the generated numerical sequence
- plot a linear trend of the generated numerical sequence
- analytically determine whether the linear trend of the generated numerical sequence increases, decreases, or remains unchanged (does not increase or decrease)
The result should be a graph displaying the numerical sequence, the linear trend, and

the conclusion - the linear trend of the generated numerical sequence decreases, increases, or remains unchanged.льности.

# Solution

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Generate a random numerical sequence of 10 elements (from 1 to 20)

```python
np.random.seed(68)  # set a random seed for reproducibility of results
sequence = np.random.randint(1, 21, 10) # generate sequence
```

## Plot a graph of the generated numerical sequence

```python
x = np.arange(1, 11) # set the X axis

# plot the sequence
plt.plot(x, # X axis
         sequence, # Y axis 
         'ko-', # k -> black color, 0 -> draw a marker at each data point, - -> draw a line betwwen each data point 
         label='Random sequence') # plot label
plt.legend() # show the plot label
```

## Plot a linear trend of the generated numerical sequence

```python
# get needed parameters for linear regression model
slope, intercept, r_value, p_value, std_err = stats.linregress(x, sequence)

# drawe a trend line using calculated parameters
trend_line = slope * x + intercept

# plot the result
plt.plot(x, trend_line, 'r--', label='Linear trend')
plt.legend() # show the plot label
```

## Analytically determine whether the linear trend of the generated numerical sequence increases, decreases, or remains unchanged

```python
if slope > 0:
    trend_direction = 'linear trend of the generated numerical sequence increasing'
    print(trend_direction)
elif slope < 0:
    trend_direction = 'linear trend of the generated numerical sequence decreasing'
    print(trend_direction)
else:
    trend_direction = 'linear trend of the generated numerical sequence stays the samee'
    print(trend_direction)
```

## Final result

```python
plt.plot(x,
         sequence,
         'ko-',
         label='Random sequence')

trend_line = slope * x + intercept
plt.plot(x, trend_line, 'r--', label='Linear trend')

plt.legend()

print(f'Наклон (slope): {slope}')
```
